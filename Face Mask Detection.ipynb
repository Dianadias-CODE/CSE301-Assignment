```python
# 1. Import Libraries
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# 2. Create Data Generator
data = ImageDataGenerator(rescale=1./255, validation_split=0.2)

# 3. Load Training Data
train = data.flow_from_directory(
    "dataset", target_size=(128,128),
    batch_size=32, class_mode="binary", subset="training")

# 4. Load Validation Data
val = data.flow_from_directory(
    "dataset", target_size=(128,128),
    batch_size=32, class_mode="binary", subset="validation")

model = tf.keras.Sequential([
    # 1. Convolutional Layer
    tf.keras.layers.Conv2D(32, 3, activation='relu', input_shape=(128,128,3)),
    # This layer applies 32 filters (3x3) to the input image.
    # It detects features like edges, textures, or patterns.
    # 'relu' introduces non-linearity so the network can learn complex patterns.

    # 2. Max Pooling Layer
    tf.keras.layers.MaxPooling2D(),
    # This layer downsamples the feature maps by taking the maximum value in each 2x2 window.
    # It reduces computation and keeps important features.

  # 3. Flatten Layer
    tf.keras.layers.Flatten(),
    # Converts the 2D feature maps into a 1D vector so it can be fed into the Dense layer.

    # 4. Dense (Fully Connected) Layer
    tf.keras.layers.Dense(1, activation='sigmoid')
    # This is the output layer for binary classification.
    # 'sigmoid' gives a probability between 0 and 1 for the two classes.
])

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

model.fit(train, validation_data=val, epochs=5)

```
